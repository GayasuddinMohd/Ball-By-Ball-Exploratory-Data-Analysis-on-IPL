# Which bowler takes the most wickets in the match 1 ?

```python
# Dataset where batsman get bowled or caught out by the bowler in the match 1.
bowler=match_1.loc[(match_1['dismissal_kind']=='caught') | (match_1['dismissal_kind']=='bowled')]
bowler
```

```python
# Names of the bowlers and their respective number of wickets.
bowler['bowler'].value_counts()
```

```python
# List of names of bowlers who takes most wickets in the match 1.
list(bowler['bowler'].value_counts().keys())
```

```python
# List of the number of most wickets taken by the bowler in the match 1.
list(bowler['bowler'].value_counts())
```

```python
# Count-plot of top wicket takers bowlers in the match 1.
plt.figure(figsize=(11,5))
sns.countplot(x='bowler',order=bowler['bowler'].value_counts().index[0:9],data=bowler,palette = "Set1",ec='black',lw=1)
plt.title('Wickets in the Match 1',fontsize=15)
plt.grid(True)
plt.show()
```
